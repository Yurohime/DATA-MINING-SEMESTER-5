## Machine Learning

1. Splitting X and y

```python
X = df.drop('income', axis=1)
y = df['income']
```

2. Split data into training and test set (test size 20% and random state 42) and check the shape of X_train and X_test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape, X_test.shape
```

## Desicion Tree

b. Predict the test set results <br>

1. Criterion Entropy

```python
# import library
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# a. Define Model
clf_en = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
clf_en.fit(X_train, y_train)
```

```python
# b. Predict the test set results
y_pred_en = clf_en.predict(X_test)
```

```python
# c. Check Accuracy Score
print("Model accuracy score with criterion entropy: {0:0.4f}".format(accuracy_score(y_test, y_pred_en)))
```

```python
# d. Check for overfitting and underfitting
print("Training set score: {:0.4f}".format(clf_en.score(X_train, y_train)))
print("Test set score: {:0.4f}".format(clf_en.score(X_test, y_test)))
```

```python
# e. Visualize decision trees
plt.figure(figsize=(12,8))
tree.plot_tree(clf_en.fit(X_train, y_train))
```

2. Criterion Gini index

```python
# a. Define Model
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)
clf_gini.fit(X_train, y_train)
```

```python
# b. Predict the test set results
y_pred_gini = clf_gini.predict(X_test)
```

```python
# c. Check Accuracy Score
print("Model accuracy score with gini index: {0:0.4f}".format(accuracy_score(y_test, y_pred_gini)))
```

```python
# d. Check for overfitting and underfitting
print("Training set score: {:0.4f}".format(clf_gini.score(X_train, y_train)))
print("Test set score: {:0.4f}".format(clf_gini.score(X_test, y_test)))
```

```python
# e. Visualize decision trees
plt.figure(figsize=(12,8))
tree.plot_tree(clf_gini.fit(X_train, y_train))
```

3. Confusion matrix and classification report of criterion Entropy and Gini index

```python
# f. Confusion Matrix

print("Entropy")
cm = confusion_matrix(y_test, y_pred_en)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Entrophy Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

print()
print("Gini Index")

cm = confusion_matrix(y_test, y_pred_gini)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("GINI Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# g. Classification Report
print("Entropy")
print(classification_report(y_test, y_pred_en))
print()
print("Gini Index")
print(classification_report(y_test, y_pred_gini))
```

```python
# h. Results and Conclusion
```

The Entropy and Gini Index models show similar performance in classifying two classes. Both models have good precision and recall for Class 0 but lower for Class 1. The overall accuracy is around 76-77%, with macro F1-scores of 0.73 to 0.74, indicating slightly better performance in predicting the majority class.

## k-nearest neighbors algorithm

---

a. Defining the model <br>
b. Predict the test set results <br>
c. Check accuracy score <br>
d. Confusion matrix <br>
e. Classification report <br>
f. Results and conclusion <br>

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')
classifier.fit(X_train, y_train) 
y_pred = classifier.predict(X_test)
```

```python
print("Model accuracy score with KNN: {0:0.4f}".format(accuracy_score(y_test, y_pred)))
```

```python
# Display the accuracy score
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# Predict y_pred from classification model using X_test
y_pred = classifier.predict(X_test)

# Make a dataframe to compare y_test and y_pred side-by-side
comparison_table = pd.DataFrame({"y_test":y_test, "y_pred":y_pred})
comparison_table.head()

from sklearn.metrics import confusion_matrix, classification_report

cr = classification_report(y_true=y_test, y_pred=y_pred)
print("Classification Report:\n", cr)
```

```python
# f. Results and conclusion
```

The model's overall accuracy is 70%, with better performance for Class 0 than Class 1. Class 0 has higher precision, recall, and F1-score compared to Class 1, which has lower scores across the board. The macro and weighted averages for precision, recall, and F1-score hover around 0.68 to 0.69, indicating moderate performance, with some imbalance favoring Class 0.
