```python
# g
# Average physical attack for Fighters: 119.96551724137932
# Average physical attack for Tanks: 119.53333333333333
# Fighters have higher average physical attack than Tanks.





df_f = df.loc[df['role'] == "fighter"]
df_t = df.loc[df['role'] == "tank"]

df_f.head()

fighter_atk = df_f['physical_atk'].sum()
fighter_ttl = df_f['physical_atk'].count()

fighter_avg = fighter_atk/fighter_ttl

tank_atk = df_t['physical_atk'].sum()
tank_ttl = df_t['physical_atk'].count()

tank_avg = tank_atk/tank_ttl

print("Average physical attack for Fighters: "+str(fighter_avg))
print("Average physical attack for Tanks: "+str(tank_avg))
if (fighter_avg > tank_avg):
    print("Fighters have higher average physical attack than Tanks")
else:
    print("Tanks have higher average physical attack than Fighters")

```

7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import metrics 

    
# Encode the 'role' column
df['role'] = df['role'].map({'fighter': 0, 'tank': 1, 'support': 2, 'marksman': 3, 'mage': 4,  'assassin': 5})
df.head()
```

```python

# Separate features and target variable
y = df['role']
X = df[[
    'defense_overall',
    'offense_overall',	
    'skill_effect_overall',
    'difficulty_overall',
    'movement_spd',
    'magic_defense',
    'mana',	
    'hp_regen',
    'physical_atk',
    'physical_defense',
    'hp',
    'attack_speed',
    'mana_regen']]



# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train the model
clf = clf.fit(X_train,y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# Predict the role for the given character
prediction = clf.predict([[6,7,4,3,250,9,380,37,119,24,2689,0.75,16.0]])

# Get the predicted role (dicoding)
predicted_role = ""
if prediction == 0:
    predicted_role = "Fighter"
elif prediction == 1:
    predicted_role = "tank"
elif prediction == 2:
    predicted_role = "Support"
elif prediction == 3:
    predicted_role = "Marksman"
elif prediction == 4:
    predicted_role = "Mage"
elif prediction == 5:
    predicted_role = "Assassin"
else:
    predicted_role = "Unknown"


print("The Predicted Role for this guy is a: "+str(predicted_role))
```
